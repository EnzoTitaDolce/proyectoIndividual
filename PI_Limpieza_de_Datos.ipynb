```python
#hacmeos las imporaciones necesarias
import pandas as pd
import numpy as np
from pandas import json_normalize #para convertir diccionarios y listas
import ast #para convettir strings en listas o diccionarios
```

```python
#configuramos pandas para que muestre todas las columnas del dataframe
pd.set_option('display.max_columns',None)

#importamos el archivo csv
datos=pd.read_csv('../../PI MLOps-20230726T005937Z-001/PI MLOps/movies_dataset.csv')
```

```python
#convertimos el archivo csv en un dataframe
data=pd.DataFrame(datos)

#vemos cuantas columnas tiene y la cantidad de valores no nulos del datafram
data.info()
```

```python
"completamos los valores faltantes dela columna revenue con el valor 0"
data['revenue'].fillna(0,inplace=True)

#volvemos a imprimir un resumen del dataframe para ver si la cantidad de valores nulos de la columna revenue cambi√≥
data.info()
```

```python
#idem que con la columna revenue pero con la columna budget
data['budget'].fillna(0, inplace=True)
data.head()
```

```python
#eliminamos los valores faltantes pero solo de la columna release_date
data.dropna(subset=['release_date'],inplace=True)
```

```python
data.info()
```

```python
#hacemos una copia del dataset para probar el desanidamiento de datos
data2=data.copy()

#eliminamos las columnas innecesarias
eliminar=['video','imdb_id','adult','original_title','poster_path','homepage']
data2.drop(columns=eliminar, inplace=True)
data2
```

```python
#la columna belongs_to_collection parece ser un diccionario pero es en realidad un string
type(data2['belongs_to_collection'][0])
```

```python
#para convertira en diccionario usamos la librer√≠a ast
#los valores nulos en la columna belong_to_collection ocasionan un error en el m√©todo literal_eval
#para evitar eso los completamos el m√©todo fillna con una lista vac√≠a
data2['belongs_to_collection'].fillna('[]',inplace=True)
data2['belongs_to_collection']=data2['belongs_to_collection'].apply(ast.literal_eval)
```

```python
#idem con las otras columnas, primero completamos los valores faltantes con listas vac√≠as
data2['genres'].fillna('[]',inplace=True)
data2['production_companies'].fillna('[]',inplace=True)
data2['production_countries'].fillna('[]',inplace=True)
data2['spoken_languages'].fillna('[]',inplace=True)
```

```python
#aplicamos el m√©todo literal_eval a las columnas necesarias
columnas=['genres','production_companies','production_countries','spoken_languages']

data2[columnas]=data2[columnas].applymap(ast.literal_eval)
data2
```

```python
#verificamos que las columnas anteriores hayan cambiado su tipo de dato
type(data2['genres'][0])
```

```python
#de la columna genre nos interesa el nombre del g√©nero m√°s que su id
#prodcedemos a extraer el nombre del g√©nero de cada diccionario y a almacenarlo en una lista
#definimos la funci√≥n
def extract_genre(genrelist):
    listageneros=[genre['name'] for genre in genrelist]
    return listageneros

#le aplicamos la funci√≥n a la columna genres y guardamos la lista en una nueva columna

data2['genre_names']=data2['genres'].apply(extract_genre)
```

```python
#la operaci√≥n anterior funcion√≥
#aplicamos la misma idea a las otras columnas
data2[['belongs_to_collection','production_companies','production_countries','spoken_languages']]

#vemos que todas las columnas son diccionarios o listas de diccionarios y el atributo que nos interesa est√° guardado en la key name
```

```python
#definimos una funci√≥n para extraer el valor de las keys en caso de que las columnas sean diccionarios
def extract_dicts(diccionario):
    if type(diccionario)==dict:
        names=[diccionario['name']]
        return names    
    else:
        return []
    
#funci√≥n para el caso en que el registro sea una lista de datos
def extract_list(lista):
    if type(lista)==list:
        names=[elemento['name'] for elemento in lista]
        return names
    else:
        return []
```

```python
#aplicamos la funci√≥n a las columnas correspondientes

data2['belongs_to_collection']=data2['belongs_to_collection'].apply(extract_dicts)
data2['production_companies']=data2['production_companies'].apply(extract_list)
data2['production_countries']=data2['production_countries'].apply(extract_list)
data2['spoken_languages']=data2['spoken_languages'].apply(extract_list)
data2
```

```python
#extraemos el a√±o de la columna release_date
data2['release_year']=data2['release_date'].str[:4]
```

```python
#convertimos los string de la columna release_date a objetos del tipo date
data2['date']=pd.to_datetime(data2['release_date'],errors='coerce').dt.date
```

```python
#eliminamos las columnas innecesarias
data2.drop('release_date',axis=1, inplace=True)
```

```python
data2.drop('genres',axis=1,inplace=True)
```

```python
#intentamos convertir la columna budget a float para poder calcular la columna return pero recibimos un error de valores imposibles de convertir
data2['budget']=data2['budget'].astype(float)
```

```python
#usamos el m√©todo to_numeric de pandas para convertir la columna budget a float y reemplazar los valores inv√°lidos por 0
data2['budget']=pd.to_numeric(data['budget'],errors='coerce').fillna(0).astype(float)
```
